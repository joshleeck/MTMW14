## Description of model

Ocean gyres and the Gulf Stream are important oceanic features which occur as a response to wind stress. The Gulf Stream is an example of a western boundary current in the Atlantic Ocean. In order to understand these features, a closed ocean basin subject to wind stress can be modelled. The model of Stommel (1948) is the simplest dynamical model able to represent a wind-driven circulation in a closed ocean basin, including a western boundary current. This model is described by the shallow water equations, linearised about a resting state, with linear drag and wind stress. The equations are listed below:

$$\frac{\partial \eta}{\partial t} + H \nabla . \mathbf{V}  \qquad (1)$$
$$\frac{\partial \mathbf{V}}{\partial t} + (f_0 + \beta y)\mathbf{k} \times \mathbf{V} = -g\nabla \eta - \gamma \mathbf{V} + \frac{\mathbf{\tau}}{\rho H}     \qquad (2)$$ 

where the prognostic variables are the surface elevation $\eta (\mathbf{x},t)$ and the depth-averaged horizontal velocity $\mathbf{V}(\mathbf{x},t)$. Here, $\mathbf{x}(x, y)$ is the 2D spatial coordinate and $\mathbf{k}$ is a unit vector in the vertical.

Equations (1) and (2) are subject to no-normal flow (kinematic) boundary conditions. The computational domain is a square defined by the ranges [$0,L$] $\times$ [$0,L$].

Initially the flow is at rest and velocity ($u, v$) and elevation ($\eta$) are zero throughout the domain.

The wind stress vector $\mathbf{\tau}$ is described below:

$$\tau = \begin{pmatrix} \tau_x \\ \tau_y \end{pmatrix} = \tau_0 \begin{pmatrix} -cos(\frac{\pi y}{L}) \\ 0 \end{pmatrix}  \qquad (3)$$

The variables used in the model (assigned in <i>parameters.py</i>) are described below, along with their physical interpretation:

## Task A

According to lecture notes, the Rossby deformation radius is given by:

$$R_d = \frac{\sqrt{gH}}{f_0}  \qquad (4)$$

Using values in <i>parameters.py</i>, we have $R_d = \frac{\sqrt{10 \times 1000}}{10^{-4}} = 10^6$.

In this scenario, inertia-gravity waves are considered. These waves propagate fast and the maximum wave speed that is anticipated from these waves is $141 m s^{-1}$ in the diagonal direction ($45^{\circ}$ to the x-axis and y-axis). This suggests that we may need to use fine grid to in the model ($d << R_d$).

The threshold of $\frac{R_d}{d} > 10$ is set for the fine grid regime (from lecture notes). Using $d =$ 20000 $m$, this gives $\frac{R_d}{d} = 50$, which is above the threshold.

If the maximum wave speed is 141 $m s^{-1}$ in the diagnoal direction, this means that $u_{max} = v_{max} = 100 m s^{-1}$.

The Courant-Friedrichs-Lewy (CFL) condition in 2D is shown below:

$$ C = \frac{u \Delta t}{\Delta x} + \frac{v \Delta t}{\Delta y} = \frac{u \Delta t}{d} + \frac{v \Delta t}{d} \leq 1 \qquad (5)$$

Hence, the upper bound on the timestep to use in the model can be calculated:

$$ \Delta t \leq \frac{d}{u + v} \qquad (6)$$

The above values of $d$, $u$ and $v$ results in $\Delta t \leq \frac{20000}{100 + 100} = 141 s$. Hence, the timestep $dt$ of $140s$ is used in this model.

## Task C

The model is run for 5 periods ($nt = 6300$) without any forcing. However, non-linearity is included in the equations by setting $\epsilon = 0.1$. The effect of non-linearity on the coupled system is investigated with $\mu = \frac{2}{3}$ and $\mu = 0.75$.

```python
pm.pm['eps'] = 0.1
pm.pm['mew_0'] = 2/3.
main.main()
main.plotgraph("$\epsilon = 0.1$")
pm.pm['eps'] = 0
pm.pm['mew_0'] = 2/3.
main.main()
main.plotgraph("$\epsilon = 0$")
plt.legend(loc = 'upper right', bbox_to_anchor=(0.1,-0.2))
```

Figure 4: (a) The trajectory of the solution $h$ against $T$, and (b) the time series of $T$. Model parameters used are $b_0 = 2.5$, $\gamma = 0.75$, $c = 1$, $r = 0.25$, $\alpha = 0.125$, $\epsilon = 0.1$ (blue) and $\epsilon = 0$ (green), $\mu = \frac{2}{3}$, $\xi = 0$. When $\epsilon = 0.1$ (blue), the trajectory of the solution decays very slowly (Figure 4). This is seen in the time series as well. The amplitude of the oscillation of $T$ decreases over time. The trajectory when $\epsilon = 0$ is plotted (green) to show the comparison of the decay in the non-linear solution. Physically, the non-linearity can represent the non-linear feedback response of the SSTs on thermocline depth, and also the non-linear vertical distribution of temperature in the tropical upper ocean (Jin, 1997). This non-linearity in the system decreases growth rate of $T$ by a factor proportional to $h^3$ and $T^3$. This means that if $T$ and $h$ grows exponentially, the non-linearity will decrease the growth rate of SSTs exponentially and ensure that the system has a finite amplitude so the values of $T$ and $h$ are not unphysical.

To further analyse the effect of non-linearity on the model and illustrate this, the comparison is repeated using $\mu = 0.75$ below.

## Task D

Galanti and Tziperman (2000) introduced the annual frequency to the problem by allowing the coupling parameter, $\mu$, to vary on an annual cycle. This is given by the equation below:

$$ \mu = \mu_0[1 + \mu_{ann} cos(\frac{2\pi t}{\tau} - \frac{5\pi}{6})] \qquad (7)$$

where $\tau$ is the annual timescale, $\mu_0$ is the coupling parameter and $\mu_{ann}$ is based on the strength of the seasonal cycle. In the model setup, $\tau = 6$ (12 months after redimensionalising), $\mu_0 = 0.75$ and $\mu_{ann} = 0.2$ is used. The model is run without forcing below.

```python
pm.pm['eps'] = 0.1
pm.pm['mew_ann'] = 0.2
main.main()
main.plotgraph()
```

Figure 6: (a) The trajectory of the solution $h$ against $T$, and (b) the time series of $T$. The model is run for 6300 timesteps ($nt = 6300$) with $\Delta t = \frac{1}{60}$, which is equivalent to five periods (210 months after redimensionalising). Model parameters used are $b_0 = 2.5$, $\gamma = 0.75$, $c = 1$, $r = 0.25$, $\alpha = 0.125$, $\epsilon = 0.1$, $\mu_0 = 0.75$, $\tau = 6$, $\mu_{ann} = 0.2$, $\xi = 0$. Initial conditions of $T = 1.125$ K and $h = 0$ m are imposed.

As $\mu$ is now a function of time, the non-linearity of the system and Bjerknes feedback process characterised by $R$ becomes a function of time as well. The annual cycle is not in phase with the oscillation of the $T$, resulting in enhanced coupling or reduced coupling, and hence associated feedback processes depending on the time of the year. This influences the trajectory of the solution, and the time series of $T$. The trajectory is no longer a smooth ellipse as seen in previous model simulations, and deviates erratically from the smooth ellipse. However, the general elliptical shape in the trajectory can still be distinguished. This may be because the time variation in $\mu$ is not critical in controlling the feedback processes of the coupled system, especially since the model is uses a daily timestep ($\Delta t = \frac{1}{60} = 1$ day). The time series of $T$ still follows an oscillation with a relatively consistent period, with notably sharper peaks and troughs than previous model simulations, possibly due to the annual variability in coupling.

## Task E

Wind stress forcing is introduced to the problem by allowing $\xi$ to vary on an annual cycle. This is given by the equation below:

$$ \xi = f_{ann}cos(\frac{2\pi t}{\tau}) + f_{ran}W\frac{\tau_{cor}}{\Delta t} \qquad (8)$$

where $f_{ann}$ is based on the strength of the annual forcing cycle, $f_{ran}$ represents the amount of white noise to add to the system, $W$ represents the white noise and $\tau_{cor}$ is the white noise time interval. In the model setup, $W$ is assigned a random value assuming uniform probability from -1 to 1 (with 2 d.p.) every time $\xi$ is evaluated, $\tau_{cor} = \frac{1}{60}$ (1 day after redimensionalising), $f_{ann} = 0.02$ and $f_{ran} = 0.2$ is used. The model is run with both forcing and non-linearity below.

```python
main.main()
main.plotgraph("without wind forcing")
pm.pm['f_ann'] = 0.02
pm.pm['f_ran'] = 0.2
pm.pm['tau_cor'] = 1/60.
main.main()
main.plotgraph("with wind forcing")
plt.legend(loc = 'upper right', bbox_to_anchor=(0.4,-0.2))
```

Figure 7: (a) The trajectory of the solution $h$ against $T$, and (b) the time series of $T$. The model is run for 6300 timesteps ($nt = 6300$) with $\Delta t = \frac{1}{60}$, which is equivalent to five periods (210 months after redimensionalising). Model parameters used are $b_0 = 2.5$, $\gamma = 0.75$, $c = 1$, $r = 0.25$, $\alpha = 0.125$, $\epsilon = 0.1$, $\mu_0 = 0.75$, $\tau = 6$, $\mu_{ann} = 0.2$, $\tau_{cor} = \frac{1}{60}$. Initial conditions of $T = 1.125$ K and $h = 0$ m are imposed. Wind forcing is included, with $f_{ann} = 0.02$ and $f_{ran} = 0.2$ (green) and excluded (blue) for comparison.

The trajectory of the solution (green) follows a general elliptical shape. The time series of $T$ also follows an oscillation with a relatively consistent period, with sharper peaks and troughs than in previous model runs. However, there are additional 'ripples' seen in the trajectory when wind forcing is included compared to when it is excluded. This is more prominent closer to the peaks and troughs of the time series, where these are either damped or amplified. Small peaks in the time series when wind forcing is excluded may be amplified in the time series when wind forcing is included. To investigate the effects of annual forcing and random forcing on the solution, the model is run without annual forcing ($f_{ann} = 0$ and $f_{ran} = 0.4$), and without random forcing ($f_{ann} = 0.1$ and $f_{ran} = 0$). The effects are shown below.

```python
pm.pm['f_ann'] = 0
pm.pm['f_ran'] = 0.4
main.main()
main.plotgraph("$f_{ann}=0, f_{ran}=0.4$")
pm.pm['f_ann'] = 0.1
pm.pm['f_ran'] = 0
main.main()
main.plotgraph("$f_{ann}=0.1, f_{ran}=0$")
plt.legend(loc = 'upper right', bbox_to_anchor=(0.4,-0.2))
```

There are 'ripples' in the trajectory with random forcing component (blue) only. This suggests that the white noise introduces randomness in the solution which appear as 'ripples'. In reality, there is always noise associated with the forcing, so the random 'ripples' causes uncertainty in $T$ and $h$ but do not alter the trajectory drastically. There are also damping and amplification of peaks and troughs in the time series of $T$ introduced by the annual component of wind forcing. This suggests that the damping and amplification of peaks and troughs are due to the annual component of wind forcing. This also shows that the annual changes in winds over the Pacific Ocean are important in modulating feedback process between $T$ and $h$, especially for this model to predict how strong the ENSO event will be.

## Task F

Given the irregular nature of trajectories, the model is likely to be sensitive to initial conditions. A forecast ensemble is designed using the model to explore the ingredients necessary to produce sensitivity to initial conditions.

The initial conditions are perturbed before the start of each forecast which runs for 5 periods ($nt = 6300$), according to the estimated uncertainty in the initial state of the system. The uncertainty is assumed to be based on the errors in instrumentation and/or data assimilation, and are chosen arbitarily to demonstrate the sensitivity of the model to small perturbations. For $T$, the uncertainty is $0.1$ K (~$0.02$ when undimensionalised). For $h$, the uncertainty is $6$ m (~$0.04$ when undimensionalised). The perturbations are made on a control initial condition of $T = 1.125$ K and $h = 0$ m used in Tasks A to E. $T$ (undimensionalised) takes values of $0.13$, $0.14$, $0.16$, $0.17$ and $h$ (undimensionalised) takes values of $-0.04$, $-0.02$, $0.02$, $0.04$. This means that there are 16 ensemble members in addition to the control initial condition in the experiment setup below.

```python
pm.pm['f_ran'] = 0.2
pm.pm['f_ann'] = 0.02
pm.pm['nt'] = 6300
#control initial conditions run using T=0.15 and h=0
main.main()
main.task_f_plot(line_width=2.5)
#ensemble run with perturbed initial conditions
temp_ens = [0.13,0.14,0.16,0.17]
h_ens = [-0.04,-0.02,0.04,0.02]
for i in range(len(temp_ens)):
    pm.pm['temp0'] = temp_ens[i]
    for j in range(len(h_ens)):
        pm.pm['h0'] = h_ens[j]
        main.main()
        main.task_f_plot()
```

The randomness arises from the white noise in equation 8, which cannot be predicted in the forecast. The ensembles attempt to quantify this uncertainty in the forecast. This can be done by using the ensemble mean to estimate the signal forecast by the model. The plot shows that even with randomness, the ensemble members all still follow a general oscillatory pattern, although some are slightly out of phase with others. This suggests that there are not enough degrees of freedom for the system to be chaotic.

## Conclusion

In this project, a finite difference numerical model is devised using the 4th-order Runge Kutta scheme to simulate the Jin's recharge oscillator to explore the effects of non-linearity and forcing on coupled systems. The model is also used to perform an ensemble forecast to understand the ingredients for sensitivity to initial conditions and the nature of the system. 

The model is realistic in representing the physical mechanisms of the system. The model response to coupling and non-linearity is captured in Tasks A to E, showing that $T$ and $h$ have a simple recharge discharge feedback mechanism which results in an oscillation with a period of 42 months. This is similar to the observed ENSO signal which has a period varying between 3 to 5 years.

The model allows an understanding of how signal and noise are equally important in producing a realistic ENSO signal. The ensemble forecasts in Task F using the model reveals that the system is non-chaotic, as expected. Randomness in the forecast simulations can be included which leads to diverging ensemble forecasts. In reality, the ENSO system is more complex and has more degrees of freedom, which leads to chaos not captured in the model.
